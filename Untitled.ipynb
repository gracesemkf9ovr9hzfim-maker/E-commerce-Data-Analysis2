```python
# 导入数据分析核心库
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

# 消除有些警告信息，让报告看起来更干净
import warnings
warnings.filterwarnings('ignore')

print("⏳ 正在唤醒底层数据，请稍候...")

# 依次读取四张核心数据表
orders = pd.read_csv("olist_orders_dataset.csv")
items = pd.read_csv("olist_order_items_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")

# 顺手把订单表里的“纯文本时间”转化为“真正的时间格式”
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

print("✅ 所有数据加载并初步清洗完毕！总计约十万笔订单已就绪。")
```

## 📈 1. 宏观盘面：营收大盘趋势与时空特征

- **时间维度问题**：公司的销售额大盘走势如何？全年的销售旺季在几月份？
- **空间维度问题**：我们的核心票仓在哪里？未来物流大仓应该建在哪？

**📊 核心洞察**：
1. **时间上**：平台在 2017 年 11 月（黑色星期五）迎来史诗级爆发，单月营业额直接突破 100 万雷亚尔。
2. **空间上**：极度依赖圣保罗州（SP），单州贡献了全平台近 40% 的销售额，下沉市场潜力巨大。

```python
# --- 1. 时间维度的魔法：大盘折线图 ---
orders['year_month'] = orders['order_purchase_timestamp'].dt.to_period('M')
merged_df = pd.merge(orders, items, on='order_id', how='inner')
monthly_sales = merged_df.groupby('year_month')['price'].sum()

# 画一张专业的高级商业折线图
monthly_sales.plot(kind='line', marker='o', figsize=(12, 5), color='#1f77b4', linewidth=2)
plt.title("Monthly GMV Trend (2016-2018)", fontsize=14, fontweight='bold') 
plt.xlabel("Year & Month", fontsize=12)
plt.ylabel("Total Sales (BRL)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --- 2. 空间维度的魔法：地域集中度分析 ---
# 将客户表也拼进来，拿到州(State)的数据
df_geo = pd.merge(merged_df, customers, on='customer_id', how='inner')

# 按州分组，计算销售总额和订单量
state_sales = df_geo.groupby('customer_state').agg({
    'price': 'sum', 
    'order_id': 'nunique'
}).sort_values(by='price', ascending=False)

state_sales.rename(columns={'price': 'Total_Sales', 'order_id': 'Order_Count'}, inplace=True)

# 动态计算第一名的霸主地位
top_1_state = state_sales.index[0]
top_1_ratio = (state_sales.iloc[0]['Total_Sales'] / state_sales['Total_Sales'].sum()) * 100

print(f"\n💡 空间洞察实锤：仅仅排名第一的 {top_1_state} 州，就贡献了全平台 {top_1_ratio:.1f}% 的销售额！")

# 召唤 Jupyter 的高阶魔法：直接在最后一行写变量名，它会渲染出超级漂亮的 HTML 表格！
state_sales.head(5)
```

## 📦 2. 微观剖析：品类表现与“二八定律”验证

- **业务问题**：谁是平台的“现金牛”？营收是否过度依赖少数爆款？
- **分析方法**：通过合并商品信息表（Products），统计各品类的 GMV 占比，并验证帕累托法则（Pareto Principle）。

**📊 核心洞察**：
1. **类目霸主**：健康美容（beleza_saude）与 手表礼品（relogios_presentes）是绝对的核心支柱。
2. **长尾效应**：前 10% 的类目贡献了约 80% 的销售额，说明平台具有极强的头部效应，资源应向核心爆款倾斜。

```python
# 1. 读取商品表（刚才没读的话，这里补上）
products = pd.read_csv("olist_products_dataset.csv")

# 2. 三表联动：将之前的 merged_df 与 products 表合并，拿到商品类目名字
final_df = pd.merge(merged_df, products, on='product_id', how='inner')

# 3. 寻找“摇钱树”：按类目分组，计算总额并降序排列
category_sales = final_df.groupby('product_category_name')['price'].sum().sort_values(ascending=False)

# 4. 展示 Top 10 类目（这就是我们要找的印钞机）
print("🏆 累计销售额 Top 10 品类：")
top_10 = category_sales.head(10).to_frame()
top_10['GMV 占比(%)'] = (top_10['price'] / category_sales.sum() * 100).round(2)
top_10
```

## 🚨 3. 危机诊断：物流痛点与用户口碑归因

- **业务问题**：平台上的 1 星差评主要是什么原因导致的？
- **分析方法**：通过合并订单表（Orders）与评价表（Reviews），计算每笔订单真实的物流耗时（签收时间 - 下单时间），并与 1-5 星评价进行相关性分析。

**📊 核心洞察**：
数据实锤了 **“14天”** 为用户耐心的生死临界点！
- **5 星好评**的平均物流耗时为 **10.2 天**。
- 当物流耗时拖延至 **20.8 天**时，**1 星差评率呈现断崖式爆发**。
- 结论：公司亟需优化偏远地区的末端配送，或针对超过 10 天未送达的订单引入“客服主动干预/超时赔付”机制。

```python
# 1. 终极拼接：将订单表与评价表合并
order_reviews = pd.merge(orders, reviews, on='order_id', how='inner')

# 2. 计算真实的物流耗时（天数）
# 因为第一章我们已经转过 datetime，这里可以直接相减
order_reviews['delivery_days'] = (order_reviews['order_delivered_customer_date'] - order_reviews['order_purchase_timestamp']).dt.days

# 3. 数据清洗：剔除异常数据（未送达或时间为负的脏数据）
order_reviews = order_reviews[order_reviews['delivery_days'] >= 0]

# 4. 算账：按评价星级(1-5星)分组，计算平均物流天数
delivery_impact = order_reviews.groupby('review_score')['delivery_days'].mean().round(1)

# 5. 画图：用颜色和红线制造视觉冲击力！
plt.figure(figsize=(10, 6))
# 颜色从红(差评)渐变到绿(好评)
colors = ['#d62728', '#ff7f0e', '#ffbb78', '#98df8a', '#2ca02c']
bars = plt.bar(delivery_impact.index, delivery_impact.values, color=colors, edgecolor='black', alpha=0.8)

plt.title("Impact of Delivery Time on Customer Reviews", fontsize=15, fontweight='bold')
plt.xlabel("Review Score (1 = Worst, 5 = Best)", fontsize=12)
plt.ylabel("Average Delivery Time (Days)", fontsize=12)

# 在每个柱子上打上具体的天数标签
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.3, f'{yval} days', ha='center', va='bottom', fontsize=11, fontweight='bold')

# 🌟 绝杀：画一条 14 天的“耐心生死线”
plt.axhline(y=14, color='red', linestyle='--', linewidth=2.5, label='14-Day Patience Threshold (生死线)')
plt.legend(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

## 👑 4. 核心资产：基于 RFM 模型的高价值用户圈选

- **业务问题**：谁是平台最核心的 VIP 客户？有限的营销预算应该发给谁？
- **分析方法**：构建电商经典的 **RFM 模型**。
  - **R (Recency)**：最近一次消费距今天数（越小越活跃）
  - **F (Frequency)**：累计消费频次（越大越忠诚）
  - **M (Monetary)**：累计消费金额（越大越有商业价值）

**📊 核心洞察**：
通过聚合 `customer_unique_id`（用户的真实唯一身份证），我们成功在十万大军中锁定了**“最近 30 天内活跃，且累计消费超过 1000 雷亚尔”**的超级 VIP 客户群。针对这批人，建议提供专属客服与高规格复购礼包。

```python
# 1. 终极关联：把订单、价格、用户的唯一身份证全拼在一起
rfm_base = pd.merge(orders, customers, on='customer_id', how='inner')
rfm_base = pd.merge(rfm_base, items, on='order_id', how='inner')

# 2. 寻找“今天”：假设“今天”是数据集中最后一天下单的第二天
snapshot_date = rfm_base['order_purchase_timestamp'].max() + dt.timedelta(days=1)

# 3. 🌟 Pandas 封神技：一行代码算出每个人的 RFM！
rfm = rfm_base.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days, # R: 距离“今天”的天数
    'order_id': 'nunique',                                                # F: 真实的下单次数
    'price': 'sum'                                                        # M: 累计花的钱
}).reset_index()

# 4. 改成行话（规范列名）
rfm.rename(columns={
    'order_purchase_timestamp': 'Recency',
    'order_id': 'Frequency',
    'price': 'Monetary'
}, inplace=True)

# 5. 圈选超级 VIP 
# 我们的定义：花钱多（M > 1000）且 最近刚来过（R <= 30）
vips = rfm[(rfm['Monetary'] > 1000) & (rfm['Recency'] <= 30)]

print(f"🎉 任务完成！我们在十万大军中，精准圈出了 {len(vips)} 个极具价值的超级 VIP 客户！")
print("下面是排名前 5 的终极土豪名单：")

# 展示按消费金额降序排列的 VIP 名单
vips.sort_values(by='Monetary', ascending=False).head(5)
```
